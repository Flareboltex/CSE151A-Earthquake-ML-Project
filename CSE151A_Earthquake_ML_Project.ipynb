## Exploratory Data Analysis

Our dataset has 472 observations and 43 features.

As for our data distribution we see overwhelmingly more females than males, with 81% of the participants being females. The age distribution is fairly representative for age with a slight right skew which is expected given general age populations. For political orientation, a significant portion of the data is null, but there does seem to be more left than is representative of the population. For salary, we see that 27% are below 15000 euros, which is below the median income in Italy.

For scales, we see in house floor and number of family mebers that it goes from 0 to 5, and for any number above 5 simply puts "Over 5". Additionally, for salary the incomes are in ranges of euros such as "Between €15000 and €28000"

There is a significant amount of missing data in the salary and political orientation features with 36% and 33% respectively being null.

One feature called Out of Region Employment, Out of Region, and Out of Region Earthquake describes participants who have and haven't done certain things outside of the Campania region. For our target variables, we have fear and anxiety which represent the levels of fear and anxiety reported by the participant related to earthquakes. There is also physiological symptoms and insomnia, which are symptoms that relate to seismic stress.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data = pd.read_csv('Earthquake Perception Dataset.csv')
print(data.info())
print(data.describe())
```

```python
# Plot histograms for each numerical column
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
data[numeric_columns].hist(figsize=(10, 10), bins=15)
plt.suptitle('Histograms of Numeric Columns')
plt.show()
```

```python
# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data[numeric_columns].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

```python
# Plot scatter matrix for a subset of numeric columns
sample_numeric_columns = numeric_columns[:10]
sns.pairplot(data[sample_numeric_columns])
plt.suptitle('Scatter Plot Matrix for Selected Numeric Columns')
plt.show()
```

```python
# Plot count plots for categorical columns with few unique values
categorical_columns = data.select_dtypes(include=['object']).columns
for col in categorical_columns[:5]:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=data)
    plt.xticks(rotation=45)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Plot missing values
missing_data = data.isnull().mean() * 100
missing_data = missing_data[missing_data > 0]

plt.figure(figsize=(10, 6))
missing_data.sort_values().plot(kind='barh', color='salmon')
plt.title('Percentage of Missing Data by Column')
plt.xlabel('Percentage')
plt.show()
```
